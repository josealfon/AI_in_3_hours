```python
fig, axs = plt.subplots(ncols=7, nrows=2, figsize=(20, 10))
index = 0
axs = axs.flatten()
for k,v in pd_dataset.items():
    sns.distplot(v, ax=axs[index])
    index += 1
plt.tight_layout(pad=0.4, w_pad=0.5, h_pad=5.0)
```

What we see from graphs:

###  Correlation between features and targets

```python
plt.figure(figsize=(20, 10))
sns.heatmap(pd_dataset.corr().abs(),  annot=True)
```

Information from the correlation matrix:
1. TAX and RAD are highly correlated features (.91 of 1) 
2. All features that have correlation score above 0.5 with MEDV, can be used for prediction [LSTAT, INDUS, RM, TAX, NOX, PTRAIO] 

## Data preprocessing

### Get the dataset

```python
dataset = pd_dataset
```

```python
dataset.head()
```

### Features selection

We select features based on the correlation matrix

```python
features_for_prediction = ['LSTAT', 'INDUS', 'NOX', 'PTRATIO', 'RM', 'TAX', 'DIS', 'AGE']
```

```python
features = dataset[features_for_prediction]
```

```python
features.head()
```

```python
features = features.values
features
```

### Get targets

```python
targets = dataset['MEDV'].values
```

```python
targets[:100]
```

### Scale the dataset

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

#### Scale features

```python
features = scaler.fit_transform(features)
features
```

### Splitting data into the train and test parts

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(features, targets, test_size=0.2)
```

```python
X_train.shape
```

## Building Keras model

```python
from keras.models import Sequential
from keras.layers import Dense
```

### Define the model

```python
model = Sequential()
```

#### Constructing the model architecture

```python
model.add(Dense(13, input_dim=8, activation='relu'))
model.add(Dense(1,))
```

```python
model.summary()
```

### Compiling the model

```python
model.compile('adam', loss='mean_squared_error', metrics=['mse'])
```

### Training the model

```python
model.fit(X_train, y_train, batch_size=50, epochs=400, verbose=2)
```

### Model evaluation

```python
score = model.evaluate(X_test, y_test)
```

```python
print("Mean squared error for the test set is: ", score[0])
```

### Predicting house prices

```python
for house in range(len(X_test)):
    print("Real house price: ", y_test[house], " Predicted house price: ", model.predict(X_test[house].reshape(-1, 1).T)[0][0])
```
